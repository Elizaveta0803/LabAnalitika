```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('titanic (2).csv')
df.head()
```

# №1

```python
sns.boxplot(x="Sex", y="Age", data=df, palette='rainbow')
```

# №2

```python
sns.violinplot(x="Sex", y="Age", data=df, hue='Pclass', palette='rainbow')
```

# №3

```python
sns.countplot(x='Pclass', data=df);
```

# №4

```python
df1=df[['Age', 'Pclass', 'Survived']]
df1.head()
```

```python
sns.heatmap(df1.corr());
```

# №5

```python
g = sns.FacetGrid(df, col="Pclass", row="Survived")
g = g.map(plt.hist, "Age");
```

# №6

```python
df2=df[['Age', 'Pclass', 'Survived', 'Fare']]
df2.head()
```

```python
fig, ax = plt.subplots()
sns.regplot("Age", "Fare", data=df, ax=ax)
ax.set_title("Age")
plt.xlabel('Cоотношение возраста и стоимости билета')
```

# №7

```python
sns.lmplot(x='Age', y="Fare", size=4, aspect=2, data=df);
```

# №8

```python
sns.swarmplot(x='Survived', y='Age', data=df, palette="Set1", dodge=True)
```

# №9

```python
sns.pairplot(df)
```

# №10

```python
sns.jointplot(x='Pclass', y='Fare', data=df, kind='scatter')
```
